```python
pis['order_date']=pd.to_datetime(pis['order_date'])
```

```python
pis['order_time']=pd.to_datetime(pis['order_time'],format='%H:%M:%S')
```

```python
pis1=(pis.groupby(['order_date']).agg({'quantity':'sum'})).reset_index()
```

```python
pis1
```

```python
pd.set_option('display.max_rows', None)
```

```python
pis1['rolling_mean']=pis1['quantity'].rolling(window=7).mean()
```

```python
pis1['expanding_mean']=pis1['quantity'].expanding(7).mean()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


plt.figure(figsize=(24,12))
plt.plot(pis1['order_date'],pis1['quantity'],linestyle='-', color='#117554')
plt.plot(pis1['order_date'],pis1['rolling_mean'],linestyle='-', color='red')
plt.plot(pis1['order_date'],pis1['expanding_mean'],linestyle='-', color='blue',marker='x')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.xticks(rotation=90)
```

```python
from statsmodels.tsa.arima.model import ARIMA 

model=ARIMA(pis1['quantity'],order=(1,1,1))
model_fit=model.fit()
model_fit.forecast(steps=7)
```

```python

plt.figure(figsize=(24,12))
plt.plot(model_fit.forecast(steps=7),marker='x')
plt.ylim(135,145)
```
